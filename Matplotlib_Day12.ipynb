**Matplotlib style**<br>
There are various built-in styles in style package, and we can also write customized style files and, then, to use those styles all you need to import them and apply on the graphs and plots.In this way, we need not write various lines of code for each plot individually again and again i.e. the code is reusable whenever required. 

To display all the style

```python
import matplotlib.pyplot as plt
from matplotlib import style
print(plt.style.available)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
  
data = np.random.randn(60)
plt.style.use('dark_background')
plt.plot(data)
plt.show()
plt.style.use('Solarize_Light2')
plt.plot(data)
plt.show()
```

These styles can easily be combined by passing a list of styles.

```python
 import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
  
data = np.random.randn(60)  #random 60 values
plt.style.use([ 'Solarize_Light2','seaborn-whitegrid'])
plt.plot(data)
plt.show()
```

**Temporary styling** - If you only want to use a style for a specific block of code but don't want to change the global styling, the style package provides a context manager for limiting your changes to a specific scope.

```python
with plt.style.context('bmh'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)))  
plt.show()
```

**Line style** - The default linestyle while plotting data is solid linestyle in matplotlib. We can change this linestyle by using linestyle or ls argument of plot() method.
'solid' (default)	'-'	,'dotted'	':'	,'dashed'	'--'	,'dashdot'	'-.' <br>
**Marker reference**'o'	Circle	'*'	Star	'.'	Point	's'	Square	'D'	Diamond   etc.


```python
with plt.style.context('bmh'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'c-s')   #c-s = cyan color dashed line with square marker
plt.show()
```

**SubPlots**<br>
Matplotlib’s pyplot API has a convenience function called subplots() which acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call.We can draw multiple plots in one figure.<br>
The subplots() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.The third argument represents the index of the current plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.title('plot-1')
plt.plot(x,y)

x = np.arange(1,5)
plt.subplot(2, 2, 2)
plt.title('plot-2 Square Root')
plt.plot(x,np.sqrt(x))

plt.subplot(2, 2, 3)
plt.title('plot-3 Exponent')
plt.plot(x,np.exp(x))

plt.subplot(2, 2, 4)
plt.title('plot-4 Square')
plt.plot(x,x*x)

plt.suptitle("PLOTS")
plt.show()
```

**Multiple Plots**<br>Using Subplot function 

```python
import matplotlib.pyplot as plt
import numpy as np
import math

X = np.arange(0, math.pi*2, 0.05)  #to get angles(0 to 2pie)
figure, axis = plt.subplots(2, 2)   # creates a subplot of 2 rows by 2 columns and display 4 different plots in each subplot.

axis[0, 0].plot(X, np.sin(X))
axis[0, 0].set_title("Sine Function")
axis[0, 1].plot(X, np.cos(X))
axis[0, 1].set_title("Cosine Function")
axis[1, 0].plot(X, np.tan(X))
axis[1, 0].set_title("Tangent Function")
axis[1, 1].plot(X, np.cosh(X))
axis[1, 1].set_title("Cosh Function")
plt.show()
```

**Using subplot2grid()**<br>
This function gives more flexibility in creating an axes object at a specific location of the grid. It also allows the axes object to be spanned across multiple rows or columns.

Plt.subplot2grid(shape, location, rowspan, colspan)

```python
import matplotlib.pyplot as plt
import numpy as np
import math
X = np.arange(0, math.pi*2, 0.05)  #to get angles(0 to 2pie)
plot1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
plot2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, colspan=2)
plot3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
plot1.plot(X, np.sin(X))
plot1.set_title("Sine Function")
plot3.plot(X, np.cos(X))
plot3.set_title("Cosine Function")
plot2.plot(X, np.tan(X))
plot2.set_title("Tangent Function")
plt.show()

  
```

**Multiple Plot in same plot**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

X = np.arange(0, math.pi*2, 0.05)
plt.plot(X, np.sin(X),'c--', label='sin')
plt.plot(X, np.cos(X),'g:', label='cos')

plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine functions")

plt.legend()
plt.show()
```
