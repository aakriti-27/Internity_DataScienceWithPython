**plotly** is an interactive visualization Python library which is used to design graphs, especially interactive graphs. It can plot various graphs and charts like histogram, barplot etc It is mainly used in data analysis as well as financial analysis.<br>
**Cufflinks** is another library that connects the Pandas data frame with Plotly enabling users to create visualizations directly from Pandas.

```python
#To install plotly - pip install plotly
#To install cufflinks - pip install cufflinks

import plotly
import cufflinks as cf
import pandas as pd
import numpy as np

#Enabling the offline mode for interactive plotting locally
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()

#To display the plots
%matplotlib inline

df = pd.DataFrame(np.random.randn(100, 3), columns =['C1','C2','C3'])
df
```

```python
df.iplot()   #normal plot
```

```python
df.iplot(kind = 'bar', x = 'C1', y = 'C2')  #bar graph
```

```python
df[['C1','C3']].iplot(kind = 'scatter' , mode = 'markers')  #scatter plot
```
