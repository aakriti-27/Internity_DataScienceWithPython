**Seaborn** is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.

```python
import seaborn as sb
sb.get_dataset_names()
```

```python
df = sb.load_dataset('tips')
df
```

**Distribution plot**-This plot is used for examining univariate(one variable quantity) and bivariate distributions<br>
**Histogram**A histogram is basically used to represent data provided in a form of some groups.It is accurate method for the graphical representation of numerical data distribution.

```python
import matplotlib.pyplot as plt
sb.histplot(x='day', y='tip', data=df)
plt.show()
```

**Displot**
It is used basically for univariant set of observations and visualizes it through a histogram.

```python
sb.distplot(df['tip'], kde = False,bins=30)
#we see most people gives tip=2
```

**Joinplot**
It is used to draw a plot of two variables with bivariate and univariate graphs. It basically combines two different plots.

```python
sb.jointplot(x ='day', y ='tip', data = df)
#here we see that maximum amount of tip was received on Saturday 
```

**Pairplot**
It represents pairwise relation across the entire dataframe and supports an additional argument called hue for categorical separation.

```python
sb.pairplot(df, hue ='time')
plt.show()
```

**Rugplot**
It plots datapoints in an array as sticks on an axis.For univariant set.

```python
sb.rugplot(data=df, x="total_bill")
plt.show()
```

**Kernel Density Estimate**(KDE)plot depicts the probability density at different values in a continuous variable.

```python
sb.distplot(df['tip'],kde = True)
plt.show()
```

**Barplot**
A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean.

```python
# plot the graph using the default estimator mean
sb.barplot(x ='time', y ='total_bill', data = df)
#This shows avg total bill is more during dinner time than lunch time
```

**Countplot**
A countplot basically counts the categories and returns a count of their occurrences.

```python
sb.countplot(x ='day', data = df)
plt.show()
#This shows most of the customers visited on Saturday
```

**Boxplot**
A boxplot shows the distribution of the quantitative data that represents the comparisons between variables. boxplot shows the quartiles of the dataset. the dots indicate the presence of outliers.

```python
sb.boxplot(x ='day', y ='total_bill', data = df, hue ='sex')  #hue parameter is used to further add a categorical separation
#This plot shows males had more total bill on Sunday than females
```

**Violinplot**
Similar to the boxplot.It provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution. 

```python
sb.violinplot(x ='day', y ='total_bill', data = df, hue ='sex')
plt.plot()
```

**Stripplot**
It basically creates a scatter plot based on the category.

```python
sb.stripplot(x ='day', y ='total_bill', data = df,jitter = False)
#jitter can be used to provide displacements along the horizontal axis, which is useful when there are large clusters of data points
```

**Swarmplot**
The points are adjusted so that they do not overlap.

```python
sb.swarmplot(x ='day', y ='total_bill', data = df,hue ='smoker')
```

**Factorplot** provides a parameter called kind to choose the kind of plot

```python
sb.factorplot(x ='time', y ='tip', data = df, kind ='bar')
```

**Heatmap** is a way to show some sort of matrix plot. To use a heatmap the data should be in a matrix form. By matrix we mean that the index name and the column name must match in some way so that the data that we fill inside the cells are relevant

```python
#correlation between the different parameters
tc = df.corr()
tc
#plot a heatmap of the correlated data
sb.heatmap(tc)
plt.show()
#to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred.
```

**Pair Plot**
Pair Plot is like an automated joint plot for the entire dataset.

```python
sb.pairplot(data = df)
sb
```

**Regression plots** as the name suggests creates a regression line between 2 parameters and helps to visualize their linear relationships.<br>lmplot() makes a very simple linear regression plot.It creates a scatter plot with a linear fit on top of it.

```python
import seaborn as sb
df = sb.load_dataset('tips')
sb.lmplot(x ='total_bill', y ='tip', data = df)
```

**set_style()** function is used for plot styling.

```python
sb.set_style('dark')
sb.lmplot(x ='total_bill', y ='tip', data = df)
```

Building color palettes using the function color_palette().

```python
import matplotlib.pyplot as plt
# Set the palette using the name of a palette:
sb.set_palette("Pastel1")
sb.stripplot(x="day", y="total_bill", data=df)
plt.show()
sb.set_palette("Paired")
sb.stripplot(x="day", y="total_bill", data=df)
plt.show()
```
